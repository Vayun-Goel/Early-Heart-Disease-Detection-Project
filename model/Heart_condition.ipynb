```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="/content/drive/MyDrive/CVD_cleaned.csv"
df=pd.read_csv(path)
```

```python
df.head(5)
```

```python
import seaborn as sns
```

```python
gen_health=df["General_Health"].unique()
print(gen_health)
```

```python
df["General_Health"].mask(df["General_Health"]=="Poor",0,inplace=True)
df["General_Health"].mask(df["General_Health"]=="Fair",1,inplace=True)
df["General_Health"].mask(df["General_Health"]=="Good",2,inplace=True)
df["General_Health"].mask(df["General_Health"]=="Very Good",3,inplace=True)
df["General_Health"].mask(df["General_Health"]=="Excellent",4,inplace=True)
```

```python
df.head(5)
checkup=df["Checkup"].unique()
print(checkup)
```

```python
df["Checkup"].mask(df["Checkup"]=="Never",0,inplace=True)
df["Checkup"].mask(df["Checkup"]=="5 or more years ago",1,inplace=True)
df["Checkup"].mask(df["Checkup"]=="Within the past 5 years",2,inplace=True)
df["Checkup"].mask(df["Checkup"]=="Within the past 2 years",3,inplace=True)
df["Checkup"].mask(df["Checkup"]=="Within the past year",4,inplace=True)
df.head()
```

```python
df["Exercise"] = np.where(df["Exercise"] == "No", 0, 1)
df["Heart_Disease"] = np.where(df["Heart_Disease"] == "No", 0, 1)
df["Skin_Cancer"] = np.where(df["Skin_Cancer"] == "No", 0, 1)
df["Other_Cancer"] = np.where(df["Other_Cancer"] == "No", 0, 1)
df["Depression"] = np.where(df["Depression"] == "No", 0, 1)
df["Diabetes"] = np.where(df["Diabetes"] == "No", 0, 1)
df["Arthritis"] = np.where(df["Arthritis"] == "No", 0, 1)
df["Sex"] = np.where(df["Sex"] == "Female", 0, 1)
df["Smoking_History"] = np.where(df["Smoking_History"] == "No", 0, 1)
df.head()
```

```python
print(df["Age_Category"].unique())
```

```python
temp=df.to_numpy()
print(temp)
```

```python
for i in range(len(temp)):
  if(temp[i][10]!="80+"):
    t=(int(temp[i][10][3:])+int(temp[i][10][:2]))/2
    temp[i][10]=t
  else:
    temp[i][10]=int(85)
```
